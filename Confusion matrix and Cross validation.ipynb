# Confusion matrix
Used in classification problem.It is N x N matrix which is used to evaluate the perfomance of a classification model.N represents the number of target classes eg : _in logisitic regression with binary classes in the target will have 2x2 matrix_

  | No | Yes
-----------------
  No | TN | FP
  
  Yes | FN | TP
  
  - accuracy = TN + TP / Total:(TP + FN + FP + TP)
  - error rate: 1 - accuracy or (FP +FN)/Total
  - Precision TP/(FP +TP)
  - Recall: TP/(FN + TP)
  - FP: Type I error
  - FN: Type II error

# Cross validation
train test split not much recommended as it takes out data randomly

Cross validation wil not randomly select the training & test data, and the accuracy doesn't fluctuate(as opposed to train_test_split will vary its accuracy depending with values of some parameters: random_state, this is not desired as accuracy should depend only on the train an test data)

## Types of cross validation
### 1. LOOCV - Leave One Out CV
_eg: in a dataset with 1000 data entries, this would pick the first entry then treat the rest 999 as training data, then now pick the second entry and treat the rest 999 as training data...this continues until the last entry; no randomness here, thus in our case we would have 1000 iterations_
   - But this method takes many iterations thus more time and computational power, no.of iterations depends on numnber of data points thus larger the dataset, the more the iterations thus more computational power needed
  
### 2. K-fold CV
with the same 1000 data entries, k is the number of folds, and each fold is to have equal number of data, and:

testing data = N/K

and rest is training data

if k = 5...number of iterations = k = 5

for 1st iteration: take first 1000/5=200 as testing data and rest 800 as training data
i.e:
1. **200** | 800
2. 200 | **200** | 600
3. 200 | 200 | **200** | 400
4. 200 | 200 | 200 | **200** | 200
5. 200 | 200 | 200 | 200 | **200**

As you can see in each iteration we pick the next 200 data as test data(the one in bold)

Now in a case with **imbalance dataset** eg: target values of [0, 0, 0, 0, 0, 0, 1, 1, 2, 2]. For this case if we are to train, the k-fold would favour the zeros much than the 1's or 2's , if we do 5 k-folds , in the our last fold/iteration the test data has the last 2 data i.e : _2, 2_, but our training data has no 2, this will bring inaccuracy...**this is its disadvantage**
   

## 3. Stratified Cross validation
with our same case of 1000 entries, in which we have 600 are Yes' and 400 No's, here we solve the challenge of K-fold when we have imbalance data.We still use k as number of folds

1. **200** | 800
2. 200 | **200** | 600
3. 200 | 200 | **200** | 400
4. 200 | 200 | 200 | **200** | 200
5. 200 | 200 | 200 | 200 | **200**

in each fold, it will ensure that the number of Yes and No's in the test data(in bold) and training data are well distributed

eg in first fold we can vahe somehting like:
1. **200** _50 No and 150 Yes_ | 800 _150 No and 450 Yes_

## 4.Time series Cross validation
Deals with data that changes with time, unpredictable, same used in stock market where they collect the data, train it and now try to predict whether the curve will go up or down

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
```

```python
result
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=25)
model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
```

```python
# accuracy changes
result
```

accuracy depends on the random state, whereas the accuracy should depend on how the model predicts the actual value

Can use train_test_split in linear regression as target values are not redundant, they are within a certain range

```python
# K-fold Cross validation
from sklearn.model_selection import KFold
```

```python
k = KFold(10)  # 10 folds
```

```python
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
result = cross_val_score(model, x, y, cv=k)
```

```python
result
```

we get to see the accuracies for the 10 folds, thus know the minimum and maximum accuracy , the accuracy would be the mean accuracy of the above accuracies

```python
print("The average accuracy when k = 10:", np.mean(result))
```

```python
k = KFold(5)  # 5 folds
model = LogisticRegression()
result = cross_val_score(model, x, y, cv=k)
```

```python
print("The average accuracy when k = 5:", np.mean(result))
```

The accuracies do not differ much

```python
# Stratified K-fold CV
'''
LOOCV is no longer used due to its challnege of more iterations with larger datasets
'''

from sklearn.model_selection import StratifiedKFold
skfold = StratifiedKFold(n_splits=5)
result = cross_val_score(model, x, y, cv=skfold)
```

```python
np.mean(result)
```

```python
# LOOCV
from sklearn.model_selection import LeaveOneOut
lc = LeaveOneOut()
result = cross_val_score(model, x, y, cv=lc)
```

The LOOCV above took even more time compared to the other CVs
